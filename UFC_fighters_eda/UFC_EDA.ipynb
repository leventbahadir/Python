```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ufc_master_data.csv")
```

```python
df
```

# MOST UFC WINS BY ATHLETES (TOP 5)

```python
most_wins = df[['name', 'age', 'ufc_wins']].sort_values('ufc_wins', ascending = False).head(5)
```

```python
sns.barplot(x = 'name', y = 'ufc_wins', data = most_wins)
plt.figure(figsize=(15,8))
plt.show()

most_wins
```

I see on the top 5 that most of the top winners are above certain age. I will calculate the average age of the ufc athletes.

```python
avg_age = df[['name', 'age']]
avg_age['age'].mean() 
```

as we can see here, average age of UFC fighters is 31 years of age. There is a positive correlation between the number of fights won and the age of the fighter. Since they have a longer career, fighters get the chance of winning more fights. However, this doesnt show that they have high win ratio. In order to find the fighters with best records, I will do a small data manipulation. 

```python
most_wins = df[['name', 'age', 'ufc_wins', 'ufc_loses']].sort_values(['ufc_wins','ufc_loses'] , ascending = [False, True]).head(30)

best_records = most_wins[['name', 'age', 'ufc_wins', 'ufc_loses']].sort_values('ufc_loses', ascending = True)

best_records.head(10)
```

We see here one of the best records. These fighters hold the most wins in UFC and and has the least loosing rate out of all best winners. 

The fighter who is standing out the former P4P champion Kamaru Usman with 15 wins and 0 losses, followed by arguably the greatest of all time, Jon Jones.

# BEST FEMALE FIGHTERS

We already looked at best fighters with current records. However, UFC also has women fighters with many divisions. This part of the analysis will be focusing on the best female UFC fighters accordingly.

```python
df.drop(df.index[df['gender'] == 'Male'], inplace = True)
df.head(10)
```

I dropped the indexes with male fighters, leaving me only the female fighters for further analysis

```python
female_most_wins = df[['name', 'age', 'ufc_wins']].sort_values('ufc_wins', ascending = False).head(5)

sns.barplot(x = 'name', y = 'ufc_wins', data = female_most_wins)
plt.figure(figsize=(15,8))
plt.show()

female_most_wins
```

We see here that amanda nunes is repated two times, because she is the title holder of 2 different divisions concurrently
